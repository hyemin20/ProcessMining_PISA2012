```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
df_original = pd.read_csv('/content/drive/MyDrive/EM:IP/최종 데이터 파일/variables/0_start_with_2_original_cleaning.csv', sep=',')
df_time = pd.read_csv('/content/drive/MyDrive/EM:IP/최종 데이터 파일/variables/1_GenerateVariables_1_time_abs.csv', sep=',')
```

```python
stu_id = df_original['ID'].unique().tolist()
```

```python
events_num = []

for i in tqdm(stu_id):
    df_t = df_original[df_original['ID'] == i]    
    if max(df_t['event_number']) >= 3:
        length = max(df_t['event_number']) - 3
    else:
        length = 0 
    events_num.append(length)
```

```python
events_num
```

```python
df_time['events_num'] = events_num
```

```python
df_time['avg_time_btw_events'] = df_time['time_solving'] / df_time['events_num']
df_time
```
