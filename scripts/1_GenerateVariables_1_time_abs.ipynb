```python
# start time
time_start = []

for i in tqdm(stu_id):
    df_t = df[df['ID'] == i]
    if len(df_t) >= 3:
        start = df_t.iloc[1,6] - df_t.iloc[0,6]
    else:
        start = 0
    time_start.append(start)
```

```python
# end time
time_end = []

for i in tqdm(stu_id):
    df_t = df[df['ID'] == i]
    if len(df_t) >= 3:
        end = df_t.iloc[-1,6] - df_t.iloc[-2,6]
    else:
        end = 0
    time_end.append(end)
```

```python
# total time
time_total = []

for i in tqdm(stu_id):
    df_t = df[df['ID'] == i]
    if len(df_t) >= 2:
        total = df_t.iloc[-2,6] - df_t.iloc[0,6]
    else:
        total = 0
    time_total.append(total)
```

```python
# solving time
time_solving = []

for i in tqdm(stu_id):
    df_t = df[df['ID'] == i]
    if len(df_t) >= 4:
        solving = df_t.iloc[-2,6] - df_t.iloc[1,6]
    else:
        solving = 0
    time_solving.append(solving)
```
