```python
# start time
time_start = []

for i in tqdm(stu_id):
    df_t = df[df['ID'] == i]
    if len(df_t) > 2:
        start = df_t.iloc[1,6] - df_t.iloc[0,6]
    else:
        start = 0
    time_start.append(start)
```

```python
# end time
time_end = []

for i in tqdm(stu_id):
    df_t = df[df['ID'] == i]
    if len(df_t) > 2:
        end = df_t.iloc[-1,6] - df_t.iloc[-2,6]
    else:
        end = 0
    time_end.append(end)
```

```python
# total time
time_total = []

for i in tqdm(stu_id):
    df_t = df[df['ID'] == i]
    total = max(df_t['time'])-min(df_t['time'])
    time_total.append(total)
```
